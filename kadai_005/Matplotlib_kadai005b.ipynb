```python
import matplotlib.pyplot as plt
import numpy as np

# nrows×ncols個のサブプロットを生成する、サブプロットのレイアウトを自由にするオプション追加
fig, axes = plt.subplots(2, 3,tight_layout=True)

# 折れ線グラフのコード
y0 = [1, 2, -5, 2]

# 折れ線グラフをプロットする
axes[0,0].plot(y0)

# sin関数のコード
# 0から10までの幅で配列を等間隔に100個作成する
x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)

# sin関数をプロットする
axes[0,1].plot(x1,y1)

# ヒストグラムのコード
# 平均が15, 標準偏差5の標準正規分布から2000個の乱数の生成する
x2 = np.random .normal(15, 5, 2000)

# ヒストグラムをプロットする
axes[0,2].hist(x2)

# 散布図のコード
np.random.seed(3)
x3 = 4 + np.random.normal(0, 2, 60)
y3 = 4 + np.random.normal(0, 2, len(x3))

# 散布図をプロットする
axes[1,0].scatter(x3, y3)

# 描画
plt.show()
```
